#### 1. Implement a function that finds the factorial of a number using a loop. 

```python
n = int(input("Enter any number: "))

factorial = 1
for i in range(1, n+1):

    factorial = factorial * i
print(factorial)
```

#### 2. Write a function to calculate the average of a list of numbers provided by the user.

```python
def average(numbers):
    return sum(numbers)/len(numbers)
numbers = [10, 12, 14, 16, 18, 20]
avg = average(numbers)
print(avg)
```

#### 4.  Implement a program that takes a sentence as input and counts the number of words. 

```python
greetings = "Hi, Good morning sir"

word = 1

for i in range(len(greetings)):
    if greetings[i] == ' ':
        word+=1
print(word)
```

#### 5. Write a function to determine if a number is prime.

```python
num = int(input("Enter a number: "))
count = 0
for i in range(1, num+10):
    if num % i ==0:
        count = count + 1
if count == 2:
    print(num, "is prime number")
```
